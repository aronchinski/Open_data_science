```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('E:/Open_Data_Science_Course/Homework/01/adult.data.csv')
data.head()
```

1. Сколько мужчин и женщин (признак sex) представлено в этом наборе данных?

```python
data["sex"].value_counts()
print("Female number: {}\nMale number: {}".format(data["sex"].value_counts()[0],data["sex"].value_counts()[1]))
```

2. Каков средний возраст (признак age) женщин?

```python
t=data[data["sex"]=="Female"]["age"].mean()
print("Average Female age: {0:.2f}".format(t))
```

3. Какова доля граждан Германии (признак native-country)?


```python
t=data['native-country'].value_counts(normalize=True)["Germany"]
print("Germany as native country frequency: {0:.5f}".format(t))
```

4-5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак salary) и тех, кто получает менее 50K в год?

```python
data.groupby(['salary'])["age"].agg([np.mean, np.std])
```

6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters или Doctorate)

```python
grad_list=["Bachelors","Assoc-acdm", "Assoc-voc", "Masters", "Doctorate"]
t=all(data[data["salary"]==">50K"]["education"].apply(lambda x: x in grad_list))
if t:
    print("It's true")
else:
    print("It's false")
```

7. Выведите статистику возраста для каждой расы (признак race) и каждого пола. Используйте groupby и describe. Найдите таким образом максимальный возраст мужчин расы Amer-Indian-Eskimo.

```python
data.groupby(by=["sex","race"]).describe()
```

```python
t=data.groupby(by=["sex","race"])["age"].agg(np.max)["Male"]["Amer-Indian-Eskimo"]
print("Max age of Amer-Indian-Eskimo  Male: {}".format(t))
```

8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак marital-status)? Женатыми считаем тех, у кого marital-status начинается с Married (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.

```python
data["mar"]=data["marital-status"].apply(lambda x: True if x.startswith("Married") else False)
a=data[data["mar"]==True & (data["sex"]=="Male")]["salary"].value_counts()[">50K"]
b=data[data["mar"]==False & (data["sex"]=="Male")]["salary"].value_counts()[">50K"]
print("Married Males with salary over 50K: {}\nUnmarried Males with salary over 50K: {}".format(a,b))
data.drop("mar",axis=1)
```

9. Какое максимальное число часов человек работает в неделю (признак hours-per-week)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?

```python
t=data["hours-per-week"].agg(np.max)
print("Maximum hours-per-week: {}".format(t))
```

```python
data["mh"]=(data["hours-per-week"]==t)
pd.crosstab(data["mh"],data["salary"],margins=True)
data.drop("mh",axis=1)
```
